```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
companies = ['AAPL_data.csv','AMZN_data.csv','GOOG_data.csv','MSFT_data.csv']
df = pd.DataFrame()
for i in companies:
    d = pd.read_csv(i)
    df = pd.concat([df,d])
df.head()
```

```python
df.shape
```

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
cmpny = ['AAPL', 'AMZN', 'GOOG', 'MSFT']
```

```python
dff = df[df['Name'] == 'AAPL']
plt.plot(dff['date'],dff['close'])
```

```python
for i in cmpny:   
    dff = df[df['Name'] == i]
    plt.plot(dff['date'],dff['close'])
```

```python
plt.figure(figsize=(20,12))
for i,name in enumerate(cmpny,1):
    plt.subplot(2,2,i)
    dff = df[df['Name'] == name]
    plt.plot(dff['date'],dff['close'])
    plt.title(name)
```
